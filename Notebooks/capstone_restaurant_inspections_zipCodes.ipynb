```python
import pandas as pd

restInspectZipCodes = pd.read_csv('../data/uszips.csv')
restInspectZipCodes.head(200)
```

```python
restInspectZipCodes.drop(['lat', 'lng', 'state_name','zcta', 'parent_zcta', 'county_fips', 'county_weights', 'county_names_all', 'county_fips_all', 
                          'imprecise', 'military', 'timezone'], axis = 1, inplace = True)
restInspectZipCodes
```

```python
restInspectZipCodes['zip'] = restInspectZipCodes['zip'].astype(str)
restInspectZipCodes['zip'].dtypes
```

```python
restInspectZipCodes['population'] = restInspectZipCodes['population'].fillna(0)
restInspectZipCodes['population'] = restInspectZipCodes['population'].astype(int)
restInspectZipCodes['density'] = restInspectZipCodes['density'].fillna(0)
restInspectZipCodes['density'] = restInspectZipCodes['density'].astype(int)
```

```python
restInspectZipCodes.info()
```

```python
restInspectZipCodes.drop_duplicates(inplace=True)
```

```python
restInspectZipCodes.dropna(inplace=True)
```
